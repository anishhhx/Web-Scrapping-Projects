## Flipkart WebScrapping using BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
Names = []
Prices = []
Descriptions = []
Reviews = []
```

```python
url = "https://www.flipkart.com/search?q=iphone+under+50000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_12_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_12_sc_na_na&as-pos=2&as-type=RECENT&suggestionId=iphone+under+50000%7CMobiles&requestId=b37de46e-d65c-419f-abdd-ddccaef31dfe&as-searchtext=iphone+under+50000&page=1"
```

```python
response  = requests.get(url)
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
overall_box = soup.find("div",class_="DOjaWF gdgoEp")
```

```python
names = overall_box.find_all("div",class_="KzDlHZ")
```

```python
only_names = [ a.text.strip() for a in names]
```

```python
for b in only_names:
    Names.append(b)
```

```python
prices = overall_box.find_all("div",class_="Nx9bqj _4b5DiR")
```

```python
only_prices = [ m.text.strip() for m in prices]
```

```python
for n in only_prices:
    Prices.append(n)
```

```python
descriptions = overall_box.find_all("div",class_="_6NESgJ")
```

```python
only_descriptions = [ p.text.strip() for p in descriptions]
```

```python
for q in only_descriptions:
    Descriptions.append(q)
```

```python
reviews = overall_box.find_all("div",class_="XQDdHH")
```

```python
only_reviews = [ x.text.strip() for x in reviews]
```

```python
for y in only_reviews:
    Reviews.append(y)
```

```python
df = pd.DataFrame({'Product_Name':Names,'Product_Price':Prices,'Product_Description':Descriptions,'Product_Review':Reviews})
```

```python
csv_file = df.to_csv("Flipkart-WebScrapping.csv")
```
