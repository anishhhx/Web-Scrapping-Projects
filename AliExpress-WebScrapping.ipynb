```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.aliexpress.com/w/wholesale-smartphones-android.html?spm=a2g0o.productlist.search.0"
```

```python
web_page = requests.get(url)
web_page
```

```python
html_content = web_page.content
html_content
```

```python
soup = BeautifulSoup(html_content,"html.parser")
```

```python
print(soup.prettify())
```

```python
record = soup.find(id ="card-list")
len(record)
```

```python
name = [n.get_text() for n in record.select(".multi--titleText--nXeOvyr")]
name
```

```python
price = [p.get_text() for p in record.select(".multi--price-sale--U-S0jtj")]
price
```

```python
df = pd.DataFrame({'Name':name,'Price':price})
df
```
