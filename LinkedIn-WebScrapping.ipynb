```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.linkedin.com/jobs/search?keywords=India&location=India&geoId=102713980&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
overall_page = soup.find_all('section',class_="two-pane-serp-page__results-list")[0]
```

```python
job = overall_page.find_all('h3',class_ = "base-search-card__title")
```

```python
company = overall_page.find_all('h4',class_="base-search-card__subtitle")
```

```python
location = overall_page.find_all(class_="job-search-card__location")
```

```python
Job_Role = []
```

```python
job_role = [x.text.strip() for x in job]
```

```python
for i in job_role:
    Job_Role.append(i)
```

```python
Company = []
```

```python
company_name = [a.text.strip() for a in company]
```

```python
for b in company_name:
    Company.append(b)
```

```python
Location = []
```

```python
location_var = [m.text.strip() for m in location]
```

```python
for n in location_var:
    Location.append(n)
```

```python
df = pd.DataFrame({'Company':Company,
                   'Job_Role':Job_Role,
                   'Location':Location})
```
